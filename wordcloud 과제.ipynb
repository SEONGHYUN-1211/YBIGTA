```python
from wordcloud import WordCloud, STOPWORDS as stopwords
```

```python
# The Strange Case of Dr. Jekyll and Mr. Hyde 소설
novel = open("The Strange Case of Dr. Jekyll and Mr. Hyde.txt","r", encoding="UTF-8").read()
novel
```

```python
novelwc = WordCloud(background_color = "white", collocations = False, stopwords = stopwords)
novelwc.generate(novel)
```

```python
novelwc.words_
```

```python
# 불용어 추가적으로 제거
stopwords.update(["s", "said", "Mr"])

novelwc.generate(novel)
novelwc.to_image()
```

```python
# WordCloud 커스텀
novelwc = WordCloud(background_color = 'white',
                    collocations = False,
                    stopwords = stopwords,
                    width = 600, height = 300,
                    min_font_size = 3,
                    colormap = "plasma")

novelwc.generate(novel)
novelwc.to_image()
```

```python
import numpy as np
mask_array = np.array(Image.open("mask.png").resize((300, 300)))
mask_array
```

```python
novelwc = WordCloud(background_color = "white", collocations = False, stopwords = stopwords, mask = mask_array)

novelwc.generate(novel)
novelwc.to_image()
```
